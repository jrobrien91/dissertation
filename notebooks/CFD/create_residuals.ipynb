# CFD Solution Residual Plot

```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define Residual Directory
case = 'NASA_navyPylon_v2_tas120_aoa0_900T33'
nfile = '/Users/jrobrien/Dissertation/data/solutions/' + case + '/postProcessing/residuals/0/residuals.dat'
```

```python
df = pd.read_csv(nfile, delimiter='\t', skiprows=1)
```

```python
df
```

```python
df.keys()[0]
```

```python
fig, axarr = plt.subplots(1, 1, figsize=[12, 6])
for i in range(1, len(df.keys())):
    df.plot(x=df.keys()[0], y=df.keys()[i], ax=axarr)
axarr.semilogy(True)
axarr.set_xlabel('# of Iterations')
axarr.set_ylabel('Residual Value')
axarr.set_title(case)
plt.savefig(case + '_residual.svg')
```
