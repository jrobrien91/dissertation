# Histrogram Display of Velocity Field within CDP Sample Volume

```python
import matplotlib.pyplot as plt
```

```python
tick_label = [ ' ', '-4$^\circ$ \n Angle of Attack',  'Zero$^\circ$ \n Angle of Attack',  '+4$^\circ$ \n Angle of Attack', ' '] 
```

```python
fig, axarr = plt.subplots(2, 1, figsize=[8, 8])
axarr[0].bar(-2.25, 71.5, width=0.5, color='C1', label='Extended')
axarr[0].bar(-1.75, 115.89, width=0.5, color='C0', label='Navy')

axarr[0].bar(-0.25, 71.64, width=0.5, color='C1')
axarr[0].bar(0.25, 109.07, width=0.5, color='C0')

axarr[0].bar(1.75, 68.46, width=0.5, color='C1')
axarr[0].bar(2.25, 102.98, width=0.5, color='C0')

axarr[0].set_xlim([-3, 3])
axarr[0].set_ylim([60, 120])
# set ticks
axarr[0].set_xticks([-3, -2,  0,  2, 3])
axarr[0].set_xticklabels(tick_label, fontsize=12)
axarr[0].set_ylabel('CDP Sample Volume \n Velocity [m/s]', fontsize=16)
axarr[0].set_title('Freestream Velocity 120 [m/s]')
axarr[0].legend()

axarr[1].bar(-2.25, 82.25, width=0.5, color='C1', label='Extended')
axarr[1].bar(-1.75, 134.44, width=0.5, color='C0', label='Navy')

axarr[1].bar(-0.25, 82.54, width=0.5, color='C1')
axarr[1].bar(0.25, 126.96, width=0.5, color='C0')

axarr[1].bar(1.75, 78.87, width=0.5, color='C1')
axarr[1].bar(2.25, 119.18, width=0.5, color='C0')

axarr[1].set_xlim([-3, 3])
axarr[1].set_ylim([70, 140])
# set ticks
axarr[1].set_xticks([-3, -2,  0,  2, 3])
axarr[1].set_xticklabels(tick_label, fontsize=12)
axarr[1].set_ylabel('CDP Sample Volume \n Velocity [m/s]', fontsize=16)
axarr[1].set_title('Freestream Velocity 140 [m/s]')
axarr[1].legend()

plt.subplots_adjust(hspace=0.4)
plt.savefig('cdp_samplevolume.png')
```
